# Perceptron
- Binary Classification

```python
import numpy as np
import pandas as pd
```

## 1. Prepare training data

```python
# Training data
# study hours, pass(1) or fail(0)
training_data = np.array([
    [1, 0],
    [2, 0],
    [3, 1],
    [4, 1],
])
X = training_data[:, 0].reshape(-1, 1)  # study hours
y = training_data[:, 1]  # pass or fail
```

## 2. Initialize parameters

```python
w = 0 # weight
b = 0  # bias
learning_rate = 1
epochs = 5

print("Initial weight:", w)
print("Initial bias:", b)
```

## 3. Defining the activation function (Step Function)

```python
def step_function(x):
  return 1 if x >= 0 else 0
```

## 4. Training Loop

```python
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        # 1. Compute Weighted sum
        z = w * X[i] + b

        # 2. Apply Activation function
        y_pred = step_function(z)

        # 3. Compute Error
        error = y[i] - y_pred

        # 4. Update Weight
        w = w + learning_rate * error * X[i]

        # 5. Update Bias
        b = b + learning_rate * error

        print(f"Input: {X[i][0]}, Predicted: {y_pred}, Actual: {y[i]}, Error: {error}, Updated weight: {w[0]}, Updated bias: {b}")
```

## 5. Final Model Parameters

```python
# Final model parameters
print("Final weight:", w)
print("Final bias:", b)
```

## 6. Testing the model

```python
test_value = 3.1
z = w * test_value + b
prediction = step_function(z)
print(f"Prediction for {test_value}: {prediction}")
```
