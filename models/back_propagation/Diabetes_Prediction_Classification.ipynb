# Diabetes Risk Prediction (Binary Classification)

> *Build from scratch.*

**Neural Network Architecture:**

2 Inputs -> 3 Hidden Neurons (Sigmoid) -> 1 Output (Sigmoid)

- Result:
  - 1 -> High Risk
  - 0 -> Low Rish

```python
import numpy as np
```

```python
np.random.seed(42)

X = np.array([
    [120, 85, 150, 95], # Glucose
    [25, 30, 35, 22] # BMI
])

# Normalise values
X = X / 200

# Output : 1=High, 0=Low
y = np.array([[1, 0, 1, 0]])

print('X Shape: ', X.shape)
print('y Shape: ', y.shape)
```

## Sigmoid Activation Function

```python
# Sigmoid function
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

## Initialize Parameters

```python
# Hidden layer (3 neurons, 2 inputs)
w1 = np.random.randn(3, 2)*0.1
b1 = np.zeros((3, 1))

# Output layer (1 neuron, 3 hidden inputs)
w2 = np.random.randn(1, 3)*0.1
b2 = np.zeros((1, 1))
```

## Forward Propagation

```python
# Hidden Layer
z1 = np.dot(w1, X) + b1
a1 = sigmoid(z1)

# Output Layer
z2 = np.dot(w2, a1) + b2
a2 = sigmoid(z2)

print('a1 Shape: ', a1.shape)
print('a2 Shape: ', a2.shape)

print('\nPredicted Output:')
print(a2)
```

## Loss Function
Binary Cross Entropy

```python
m=y.shape[1]
loss =- np.sum(y*np.log(a2+1e-8)+(1-y)*np.log(1-a2+1e-8))/m
print("Loss:",loss)
```

## Backpropagation

```python
# Output layer gradient
dZ2=a2-y
dW2=(1/m)*np.dot(dZ2,a1.T)
db2=(1/m)*np.sum(dZ2,axis=1,keepdims=True)
```

## Hidden Layer Gradient

```python
# Sigmoid derivative
dA1=np.dot(w2.T,dZ2)
dZ1=dA1*a1*(1-a1)

dW1=(1/m)*np.dot(dZ1,X.T)
db1=(1/m)*np.sum(dZ1,axis=1,keepdims=True)
```

## Update Parameters

```python
learning_rate=0.5
w2 -= learning_rate*dW2
b2 -= learning_rate*db2

w1 -= learning_rate*dW1
b1 -= learning_rate*db1
```

## Complete Training Loop

```python
epochs = 5000
for i in range(epochs):
    # Forward pass
    z1 = np.dot(w1, X) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(w2, a1) + b2
    a2 = sigmoid(z2)

    # Compute loss
    loss = -np.sum(y * np.log(a2) + (1 - y) * np.log(1 - a2)) / m

    # Backward pass
    dZ2 = a2 - y
    dW2 = np.dot(dZ2, a1.T) / m
    db2 = np.sum(dZ2, axis=1, keepdims=True) / m

    dA1 = np.dot(w2.T, dZ2)
    dZ1 = dA1 * a1 * (1 - a1)

    dW1 = np.dot(dZ1, X.T) / m
    db1 = np.sum(dZ1, axis=1, keepdims=True) / m

    # Update weights and biases
    w2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    w1 -= learning_rate * dW1
    b1 -= learning_rate * db1

    if i % 500 == 0:
        print(f"Epoch {i}, Loss: {loss:.4f}")
```
